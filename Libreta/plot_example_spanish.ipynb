Lo primero de todo, es importar del fichero la función:

* `plot`: Carga un mapa desde un fichero y lo dibuja. Además, también grafica la ruta de una intersección a otra si se le pasa la ruta correspondiente. Toma como parámetros:
    * `path`: La ruta del archivo que contiene la información del mapa.
    * `route`: Una lista opcional de intersecciones que representan un camino a seguir en el mapa. Si no se proporciona, se dibuja el mapa sin ruta.

A continuación, vamos a ver un ejemplo de uso de cómo pintar únicamente el mapa:

```python
# Third party
from pathlib import Path

# Local application
from plot import plot

# Display the map based on the specified path
path = './problems/huge/calle_herreros_albacete_2000_2.json'
plot(path)
```

Una ruta desde una intersección a otra:

```python
# Display the map and the route between two intersections
route = [1255547643, 1257471997, 1529477002, 1265098200]
plot(path, route)
```

Y, lo más interesante, desde la intersección inicial a la final:

```python
# Display the map and the route between the initial and final intersections
route = [2000100687, 571129942, 1838967432, 772970904]
plot(path, route)
```

Finalmente, comentar que si la ruta no es válida, esto es, estamos tratando de pasar de una intersección a otra sin que exista un segmento que los una, aparece el siguiente error: 

```python
# Show an error because there is no route between the nodes
route = [2000100687, 955316231]
plot(path, route)
```
